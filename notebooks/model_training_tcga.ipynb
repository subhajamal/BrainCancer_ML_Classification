```python
import os
from radiomics import featureextractor
import pandas as pd

# Function to filter file paths
def filter_file_paths(rootDir, file_extensions):
    filtered_paths = []
    for dirName, subdirList, fileList in os.walk(rootDir):
        for fname in fileList:
            if any(fname.endswith(ext) for ext in file_extensions):
                filtered_paths.append(os.path.join(dirName, fname))
    return filtered_paths

# Define the root directory
rootDir = "/content/drive/MyDrive/final_project/Rembrandt_data/Rembrandt_Final_Data/"

# Define the file extensions for images
image_extensions = ['t1_LPS_rSRI.nii', 't1_LPS_rSRI.nii.gz']

# Define the file extension for label files
label_extension = 'GlistrBoost_out.nii.gz'

# Get filtered image file paths
image_paths = filter_file_paths(rootDir, image_extensions)

# Instantiate the extractor
extractor = featureextractor.RadiomicsFeatureExtractor()

# List to store results
results = []

# Extract features for each pair of image and label
for image_path in image_paths:
    # Generate label file path
    label_path = image_path.replace('t1_LPS_rSRI.nii', label_extension)

    # If label file path not found, try the alternative image extension
    if not os.path.exists(label_path):
        label_path = image_path.replace('t1_LPS_rSRI.nii.gz', label_extension)

    # Check if label file exists
    if not os.path.exists(label_path):
        print(f"Label file missing for image: {image_path}")
        continue

    # Get the folder name (ID)
    folder_name = os.path.basename(os.path.dirname(image_path))

    # Step 1: Print current image and label paths being processed
    print(f"Extracting features for:\nImage: {image_path}\nLabel: {label_path}")

    # Step 2: Extract features using RadiomicsFeatureExtractor
    result = extractor.execute(image_path, label_path)

    # Step 3: Append the result to the results list along with the folder name (ID)
    result['Folder_ID'] = folder_name
    results.append(result)
```

```python
# Convert list of dictionaries to DataFrame
df=pd.DataFrame(results)
print(df.shape)
df
# Save dataframe to CSV
df.to_csv('final_project_step1_radiomics_features.csv', index=False)
```

```python
# Convert the list of dictionaries to a DataFrame
results_df = pd.DataFrame(results)

# Define the output file path
output_file = "/content/drive/MyDrive/final_project/extracted_features.csv"

# Save the DataFrame to a CSV file
results_df.to_csv(output_file, index=False)

print("Feature extraction results saved to:", output_file)
```
