```python
import pandas as pd
# load both the pyradiomics and clinical TCGA data
pyradiomics_data = pd.read_csv('/content/drive/MyDrive/7009 Final Project/TCGA_pyradiomics/TCGA pyRadiomics recalculated/TCGA_pyradiomics_t1.csv')
clinical_data = pd.read_excel('/content/drive/MyDrive/7009 Final Project/TCGA_clinical_data/TCGA_GBM_LGG_clinical_data_for_task2_updated.xlsx')
```

```python
# Merge both the data
merged_df = pd.merge(pyradiomics_data, clinical_data, left_on="ID", right_on="Row.names")
```

```python
merged_df.head()
```

```python
from sklearn.model_selection import train_test_split

X = merged_df.drop(columns=["Disease_Type"])
y = merged_df["Disease_Type"]
```

```python
import numpy as np
# Iterate over each column and replace non-numeric data with NaN
for col in X.columns:
    if X[col].dtype == 'object':
        X[col] = np.nan

# drop columns with any NaN values
X.dropna(axis=1, inplace=True)
```

```python
# keep only the columns with numeric data types
X_numeric = X.select_dtypes(include=[np.number])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_numeric, y, test_size=0.2, random_state=42)
```

```python
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

svm_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('svc', SVC(kernel='linear', random_state=42))
])

svm_pipeline.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report

# Predictions with SVM
svm_predictions = svm_pipeline.predict(X_test)

print("SVM Classification Report:")
print(classification_report(y_test, svm_predictions))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

rf_classifier.fit(X_train, y_train)
```

```python
# Predictions with Random Forest
rf_predictions = rf_classifier.predict(X_test)

print("Random Forest Classification Report:")
print(classification_report(y_test, rf_predictions))
```

```python
# feature importances from the Random Forest model
importances = rf_classifier.feature_importances_

# DataFrame for visualization
feature_importances = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': importances
}).sort_values(by='Importance', ascending=False)

# Display the feature importances
print(feature_importances)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

# Define the parameter grid for Random Forest
param_dist_rf = {
    'n_estimators': randint(100, 500),
    'max_features': ['auto', 'sqrt'],
    'max_depth': randint(4, 20),
    'min_samples_split': randint(2, 11),
    'min_samples_leaf': randint(1, 11)
}

# Initialize RandomizedSearchCV
random_search_rf = RandomizedSearchCV(rf_classifier, param_distributions=param_dist_rf, n_iter=100, cv=5, scoring='accuracy', random_state=42)

# Perform the search
random_search_rf.fit(X_train, y_train)

# Best parameters for Random Forest
print("Best parameters for Random Forest:", random_search_rf.best_params_)
```

```python
# Evaluate the tuned Random Forest model
tuned_rf_predictions = random_search_rf.best_estimator_.predict(X_test)
print(classification_report(y_test, tuned_rf_predictions))
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid for SVM
param_grid_svm = {
    'svc__C': [0.1, 1, 10, 100],
    'svc__kernel': ['linear', 'rbf'],
    'svc__gamma': ['scale', 'auto']
}

# Initialize GridSearchCV
grid_search_svm = GridSearchCV(svm_pipeline, param_grid_svm, cv=5, scoring='accuracy')

# Perform the search
grid_search_svm.fit(X_train, y_train)

# Best parameters for SVM
print("Best parameters for SVM:", grid_search_svm.best_params_)
```

```python
# Evaluate the tuned SVM model
tuned_svm_predictions = grid_search_svm.best_estimator_.predict(X_test)
print(classification_report(y_test, tuned_svm_predictions))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

# Initialize the Gradient Boosting classifier
gb_classifier = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)

# Train the Gradient Boosting classifier
gb_classifier.fit(X_train, y_train)

# Predictions with Gradient Boosting
gb_predictions = gb_classifier.predict(X_test)
```

```python
# Evaluation report for Gradient Boosting
gb_report = classification_report(y_test, gb_predictions)
print("Gradient Boosting Classification Report:")
print(gb_report)
```

```python
# filter out 'Oligoastrocytoma' samples from X_numeric and y
X_numeric_filtered = X_numeric[~(y == 'Oligoastrocytoma')]
y_filtered = y[~(y == 'Oligoastrocytoma')]
```

```python
print("Original DataFrame size:", X_numeric.shape)
print("Filtered DataFrame size:", X_numeric_filtered.shape)
```

```python
# Splitting the filtered data into training and testing sets
X_train_filtered, X_test_filtered, y_train_filtered, y_test_filtered = train_test_split(
    X_numeric_filtered, y_filtered, test_size=0.2, random_state=42)
```

```python
# Create a pipeline that includes scaling and the SVM classifier
svm_pipeline_filtered = Pipeline([
    ('scaler', StandardScaler()),
    ('svm', SVC(kernel='linear', random_state=42))
])

# Train the pipeline on the filtered training data
svm_pipeline_filtered.fit(X_train_filtered, y_train_filtered)

# Predict on the test data
svm_predictions_filtered = svm_pipeline_filtered.predict(X_test_filtered)

# Classification report for SVM
print("SVM Classification Report:")
print(classification_report(y_test_filtered, svm_predictions_filtered))
```

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize the Random Forest classifier
rf_classifier_filtered = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the classifier on the filtered training data
rf_classifier_filtered.fit(X_train_filtered, y_train_filtered)

# Predict on the test data
rf_predictions_filtered = rf_classifier_filtered.predict(X_test_filtered)

# Classification report for Random Forest
print("Random Forest Classification Report:")
print(classification_report(y_test_filtered, rf_predictions_filtered))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

# Parameter grid for Random Forest
param_grid_rf = {
    'n_estimators': [50, 100, 200],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

rf_classifier = RandomForestClassifier(random_state=42)

# Set up the GridSearchCV
grid_search_rf = GridSearchCV(estimator=rf_classifier, param_grid=param_grid_rf, cv=5, scoring='accuracy', verbose=1, n_jobs=-1)
grid_search_rf.fit(X_train_filtered, y_train_filtered)

# Best parameters and score
print("Best parameters for Random Forest:", grid_search_rf.best_params_)
print("Best scoring for Random Forest:", grid_search_rf.best_score_)

# Evaluate using the best estimator
rf_best = grid_search_rf.best_estimator_
predictions_rf = rf_best.predict(X_test_filtered)
print("Classification Report for Random Forest:")
print(classification_report(y_test_filtered, predictions_rf))
```
